```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data.drop(["Alley", "FireplaceQu", "PoolQC", "Fence", "MiscFeature"], axis = 1, inplace = True)
test_data.drop(["Alley", "FireplaceQu", "PoolQC", "Fence", "MiscFeature"], axis = 1, inplace = True)
```

```python
col = list(train_data.columns)
col.remove("SalePrice")
print(col)
```

```python
non_cat_features = ['LotFrontage', "LotArea", "YearBuilt", "YearRemodAdd", "MasVnrArea", "BsmtFinSF1", "BsmtFinSF2",
                    "BsmtUnfSF", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "LowQualFinSF", "GrLivArea", "BsmtFullBath", 
                    "BsmtHalfBath", "FullBath", "HalfBath", "BedroomAbvGr", "KitchenAbvGr", "TotRmsAbvGrd", "Fireplaces", 
                    "GarageYrBlt", "GarageCars", "GarageArea", "WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", 
                    "ScreenPorch", "PoolArea", "MiscVal", "MoSold", "YrSold"]
```

```python
cat_features = col.copy()
# print(cat_features)
for i in non_cat_features:
#     print(i)
    cat_features.remove(i)
```

```python
print(cat_features)
```

```python
import category_encoders as ce
```

```python
target_enc = ce.TargetEncoder(cols = cat_features)
target_enc.fit(train_data[cat_features], train_data["SalePrice"])
```

```python
t1 = train_data.copy()
t1[cat_features] = target_enc.transform(train_data[cat_features])
t1.head()
```

```python
t1.fillna(t1.mean(), inplace = True)
```

```python
t2 = test_data.copy()
t2[cat_features] = target_enc.transform(t2[cat_features])
t2.head()
```

```python
t2.fillna(t2.mean(), inplace = True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X_train = t1[col]
Y_train = t1["SalePrice"]
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(X_train.copy())
```

```python
lr = LinearRegression(normalize = True).fit(x, Y_train)
```

```python
lr.score(x, Y_train)
```

```python
gbr = GradientBoostingRegressor().fit(x, Y_train)
```

```python
gbr.score(x, Y_train)
```

```python
mlp = MLPRegressor(hidden_layer_sizes = [40, 20, 10, 5], learning_rate_init = 0.1).fit(x, Y_train)
```

```python
mlp.score(x, Y_train)
```

```python
x_test = scaler.fit_transform(t2.copy())
```

```python
pred = gbr.predict(x_test)
```

```python
Id = np.array(test_data["Id"], dtype = int)
```

```python
print(len(Id))
print(len(pred))
```

```python
ans = pd.DataFrame(Id)
ans["SalePrice"] = pred
ans.columns = ["Id", "SalePrice"]
# ans["Id"].apply(lambda x : int(x))
ans.head()
```
